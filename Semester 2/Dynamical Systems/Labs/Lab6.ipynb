```python
from sage.plot.line import Line
N = 40
r = 4
x = zero_vector(RR, N + 1)
x[0] = 0.67
y = zero_vector(RR, N + 1)
y[0] = 0.67001
for n in range(1, N):
    x[n] = r*x[n - 1]*(1 - x[n-1])
    y[n] = r*y[n - 1]*(1 - y[n-1])
lx = line(Line(range(N), x, {}), color = 'blue', marker = 'o')
ly = line(Line(range(n), y, {}), color = 'red', marker = 'o')
show(lx + ly)
```

```python
from sage.plot.line import Line
N = 11
h = 1/(N - 1)
x = zero_vector(RR, N)
y = zero_vector(RR, N)
y[0] = 1
ye = zero_vector(RR, N)
ye[0] = 1
for k in range(1, N):
    x[k] = x[k - 1] + h
    y[k] = y[k - 1] + h*2*x[k - 1]*y[k - 1]
    ye[k] = exp(x[k]^2)

ly = line(Line(range(n), y, {}), color = 'red', marker = 'o')
lye = line(Line(range(n), ye, {}), color = 'blue', marker = 'o')
show(ly + lye)
```
